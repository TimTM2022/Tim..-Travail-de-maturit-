```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras import Sequential
from keras.layers import Dense
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
print(tf.__version__)
import matplotlib.pyplot as plt

```

```python
dft= pd.read_csv('Cor√©cupjupy.csv', sep=';')
del dft[dft.columns[0]]
del dft[dft.columns[147]]
dft=pd.DataFrame.to_numpy(dft)
dft=dft.astype(int)
x, y = dft[:, [0, 1]], dft[:, [2,3]]
```

```python
y=np.nan_to_num(y)
x=np.nan_to_num(x)
y
```

```python
ytrain, yval, ytest=np.split((y), [int(0.8*len(y)), int(0.9*len(y))])
print(len(yval))
print(len(ytrain))
print(len(ytest))
```

```python
train, val, test=np.split((x), [int(0.8*len(x)), int(0.9*len(x))])
```

```python
normalizer = tf.keras.layers.Normalization(axis=1)
normalizer.adapt(np.array(train))
train=normalizer(train).numpy()
ytrain=normalizer(ytrain).numpy()
ytrain=np.absolute(ytrain)
train=np.absolute(train)
```

```python
print(len(train))
print(len(val))
print(len(test))
```

```python


input1 = tf.keras.Input(shape=(2,))
l1 = tf.keras.layers.Dense(5, activation='relu')(input1)
l2=tf.keras.layers.Dense(50, activation='relu')(l1)
l3=tf.keras.layers.Dense(5, activation='relu')(l2)
l4=tf.keras.layers.Dense(5, activation='relu')(l3)
out1 = tf.keras.layers.Dense(2)(l4)
out2 = tf.keras.layers.Dense(2)(l4)
out3 = tf.keras.layers.Dense(2)(l4)
out4 = tf.keras.layers.Dense(2)(l4)
out5 = tf.keras.layers.Dense(2)(l4)


model = tf.keras.Model(inputs=input1, outputs=(out1, out2, out3, out4, out5))
```

```python
opt = keras.optimizers.Adam(learning_rate=0.0001, clipnorm=1.0)
model.compile( loss='mean_squared_error', optimizer=opt, metrics=["MSE"])
```

```python
history = model.fit(train, ytrain, batch_size=16, epochs=500, validation_data=(val, yval),verbose=1)
```

```python
test_scores = model.evaluate(test, ytest, verbose=1)
print("Test loss:", test_scores[0])
print("Test accuracy:", test_scores[1])
```
